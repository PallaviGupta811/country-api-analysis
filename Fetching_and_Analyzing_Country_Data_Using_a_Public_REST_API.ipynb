##Task 1 — Fetch and Validate Response

```python
import requests

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)
    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total countries:", len(data))
    else:
        print("Request failed")
        print(response.text)

except Exception as e:
    print("Error:", e)
```

##Task 2 — Extract and Display Data

```python
import pandas as pd

country_list = []

for country in data:
    name = country.get("name", {}).get("common", None)
    population = country.get("population", None)
    region = country.get("region", None)
    area = country.get("area", None)

    country_list.append([name, population, region, area])

df = pd.DataFrame(country_list, columns=["common_name", "population", "region", "area"])

df.head(10)
```

##Task 3 — Summarize Data

```python
# Region with highest total population
region_pop = df.groupby("region")["population"].sum()
max_region = region_pop.idxmax()

print("Region with highest total population:", max_region)

# Country with largest area
largest_country = df.loc[df["area"].idxmax(), "common_name"]

print("Country with largest area:", largest_country)
```
